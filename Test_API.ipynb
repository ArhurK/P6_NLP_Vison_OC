```python
# URL de l'API Edamam
url = 'https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser'

# Paramètres de la requête
params = {
    'nutrition-type': 'cooking',
    'ingr' : 'champagne',
    #'q': 'champagne',  # Ingrédient recherché
    'limit': 10,       # Limite de résultats
}

# En-têtes de la requête
headers = {
    'X-RapidAPI-Key': api_key,
    'X-RapidAPI-Host': 'edamam-food-and-grocery-database.p.rapidapi.com',
}

# Effectuer la requête GET
response = requests.get(url, params=params, headers=headers)

# Vérifier la réponse de l'API
if response.status_code == 200:
    data = response.json()
    # Afficher les 10 premiers produits contenant l'ingrédient "champagne"
    for i, product in enumerate(data['hints'][:10], start=1):
        print(f"{i}. {product['food']['label']}")
else:
    print("Erreur de requête à l'API :", response.status_code)
```

```python
import pandas as pd
from pandas import json_normalize

data2 = data['hints'][:10]
l = []
for i, j in enumerate(data2):
    df = json_normalize(data2[i])
    l.append(df)
df = pd.concat(l)
df = df[['food.foodId', 'food.label', 'food.category', 'food.image', 'food.foodContentsLabel']].reset_index(drop=True)
df
```

On exporte au format csv

```python
# export en csv
df.to_csv('test_api.csv')
```
